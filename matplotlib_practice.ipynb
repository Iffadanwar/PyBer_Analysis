```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot.
plt.plot(x_axis,y_axis)
plt.show()
```

```python
# changing the fig size
plt.figure(figsize=(20,10))
# Create the plot and add a label for the legend. also adding color of line, wodth of line and marker for points
plt.plot(x_axis,y_axis, label = "Boston", color="blue" , linewidth = 2, marker = "*" )
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
#set y limit 0-45
plt.ylim(0,45)
#plt.xlim("Jan","June")
#title
plt.title("PyBer Fare by Month")
#legend
plt.legend()
#adding a grid
plt.grid()
plt.show
```

```python
# changing the fig size
fig, ax = plt.subplots(figsize=(20,10))
# Create the plot and add a label for the legend. also adding color of line, wodth of line and marker for points
ax.plot(x_axis,y_axis, label = "Boston", color="blue" , linewidth = 2, marker = "*" )
# Create labels for the x and y axes.
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
#set y limit 0-45
ax.set_ylim(0,45)
#ax.set_xlim("Jan","June")
#title
ax.set_title("PyBer Fare by Month")
#legend
ax.legend()
#adding a grid
ax.grid()
```
